##Get classes and entries per class

```python
import os
import pandas as pd
```

```python
d = '/Users/sep/Documents/HAW/Vorlesung/HP/Hauptprojekt/Daten_sorted.nosync/'
PATH = '../'

classes = (os.listdir(d))
#MacOS finds a hidden file ".DS_Store". This needs to be removed from the list to avoid a index error
classes.remove('.DS_Store')
print(classes)

paths = [os.path.join(d, o) for o in os.listdir(d) 
                    if os.path.isdir(os.path.join(d,o))]

nbEntries = []

for i in range(len(classes)):
    nbEntries.append(len(os.listdir(paths[i])))

#################################################
    
print(classes)
print(nbEntries)

df = pd.DataFrame({'classes':classes, 'entries':nbEntries})
ax = df.sort_values(by='entries', ascending=True).plot.bar(x='classes', y='entries', color='cornflowerblue',legend=False, figsize=(12,8))
ax.set_title('DocILE classes Distribution')
ax.set_ylabel("number of Document images")
for p in ax.patches:
    ax.annotate(str(p.get_height()), xy=(p.get_x(), p.get_height()-30))
```
