```python
import os
import json
import shutil
import random
from tqdm.notebook import tqdm
```

Create ~150 train examples

```python
dir = "/home/sebastian/Documents/Hauptprojekt/Daten_Donut.nosync/train/"
for f in os.listdir(dir):
    os.remove(os.path.join(dir, f))

lines = []
images = []

all_images = os.listdir("/home/sebastian/Documents/Hauptprojekt/Daten.nosync/data/docile/annotations/")
random.shuffle(all_images)
counter = 0
for ann in tqdm(all_images):
    if counter == 150:
        break
    if ann != ".ipynb_checkpoints":
        with open("/home/sebastian/Documents/Hauptprojekt/Daten.nosync/data/docile/annotations/" + ann) as f:
            data = json.load(f)
    if data["metadata"]["page_count"] == 1:
        page = data["metadata"]["page_count"]
        images.append(ann[:-5] + "(" + str(page) + ")" + ".jpg")
        line = {"gt_parse": data}
        lines.append(line)
        counter+=1
#print(lines)
print(len(images))
with open("/home/sebastian/Documents/Hauptprojekt/Daten_Donut.nosync/train/metadata.jsonl", 'w') as f:
    for i, gt_parse in enumerate(lines):
        line = {"file_name": images[i], "ground_truth": json.dumps(gt_parse)}
        f.write(json.dumps(line) + "\n")
        shutil.copyfile("/home/sebastian/Documents/Hauptprojekt/Daten.nosync/data/docile/jpgs/" + images[i], "/home/sebastian/Documents/Hauptprojekt/Daten_Donut.nosync/train/" + images[i])
```
